```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # best for Colab Free

model.train(
    data=f"{dataset.location}/data.yaml",
    imgsz=640,
    epochs=50,
    batch=16,
    patience=10
)
```
